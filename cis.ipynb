```python
train_df = pd.merge(train_transaction,train_identity,on='TransactionID',how = 'left')

test_df = pd.merge(test_transation,test_identity,on='TransactionID',how = 'left')


# The merging will be done on the TransactionID feature for both of the dataframes and it will be done left.
# how = 'left' --> Means the join will happen with the TransactionID of the train_transaction because it is left. 
# There are many such values of the How parameter.
# left --> Left frame will be used.
# right --> Right frame will be used.
# Outter --> Union of the keys from both of the frames.
# Inner -->  Uses intersection of the both frames.
```

# Label Encoding

#

```python
from sklearn.preprocessing import LabelEncoder
label_encoding_cols = ['M1','M2','M3','M4','M6']

for cols in label_encoding_cols:
    encoder = LabelEncoder()
    encoder.fit(pd.concat([X_train[cols] , X_test[cols]],axis = 0))
    X_train[cols] = encoder.transform(X_train[cols])
    X_test[cols] = encoder.transform(X_test[cols])
    test[cols] = encoder.transform(test[cols])
```

```python
test
```

# One hot Encoding

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

train_index = X_train.index
test_index  = X_test.index
test_new_index = test.index
ohe_cols = ['ProductCD','card4', 'card6','DeviceType']
ohe = OneHotEncoder(drop='first',sparse_output=False , handle_unknown="ignore")
# handle parameter is used for the categories which are there in X_train and not in X_test and vice a versa

preprocessing = ColumnTransformer([('ohe',ohe,ohe_cols)],remainder='passthrough')
X_train = preprocessing.fit_transform(X_train)
X_test = preprocessing.transform(X_test)
test = preprocessing.transform(test)


feature_names = preprocessing.get_feature_names_out()

X_train = pd.DataFrame(X_train, columns=feature_names, index=train_index)
X_test  = pd.DataFrame(X_test, columns=feature_names, index=test_index)
test = pd.DataFrame(test,columns = feature_names,index = test_new_index) 
```

# Model T raining

```python
# from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.metrics import accuracy_score , classification_report , ConfusionMatrixDisplay, precision_score , f1_score , recall_score , roc_auc_score , roc_curve
# "Random_Forest_Classification" : RandomForestClassifier(),
models = {
    "Xgboost" : XGBClassifier(tree_method="hist",predictor="gpu_predictor",device = 'cuda')
}

for i in range(len(list(models))):
    model = list(models.values())[i]
    model.fit(X_train , Y_train)
    
    Y_pred_test = model.predict(X_test)
    Y_pred_train = model.predict(X_train)
    
    model_train_accuracy = accuracy_score(Y_train , Y_pred_train)
    model_train_f1 = f1_score(Y_train , Y_pred_train , average='weighted')
    model_train_precision = precision_score(Y_train,Y_pred_train)
    model_train_recall = recall_score(Y_train , Y_pred_train)
    model_train_roc_auc_score = roc_auc_score(Y_train , Y_pred_train)
    model_train_classification = classification_report(Y_train , Y_pred_train)
    
    model_test_accuracy = accuracy_score(Y_test , Y_pred_test)
    model_test_f1 = f1_score(Y_test , Y_pred_test , average='weighted')
    model_test_precision = precision_score(Y_test , Y_pred_test)
    model_test_recall = recall_score(Y_test , Y_pred_test)
    model_test_roc_auc_score = roc_auc_score(Y_test , Y_pred_test)
    model_test_classification = classification_report(Y_test , Y_pred_test)
    
    print(f"{'-'*35} {list(models.keys())[i]} {'-'*35}")
    
    print("Model Performance for training set")
    print("The Training Accuracy = {:.4f}".format(model_train_accuracy))
    print("The Training f1-score = {:.4f}".format(model_train_f1))
    print("The Training Precision = {:.4f}".format(model_train_precision))
    print("The Training ROC AUC Score = {:.4f}".format(model_train_roc_auc_score))
    print(f"{'\n'}{model_train_classification}")
    
    print("-"*75)
    
    print("Model Performance for test set")
    print("The Test Accuracy = {:.4f}".format(model_test_accuracy))
    print("The Test f1-score = {:.4f}".format(model_test_f1))
    print("The Test Precision = {:.4f}".format(model_test_precision))
    print("The Test ROC AUC Score = {:.4f}".format(model_test_roc_auc_score))
    print(f"{'\n'}{model_test_classification}")
    
    print("="*75)
    print('\n')
```
