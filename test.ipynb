```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
%matplotlib inline
warnings.filterwarnings('ignore')
```

```python
test_identity = pd.read_csv("test_identity.csv")
test_transaction = pd.read_csv("test_transaction.csv")

test_df = pd.merge(test_transaction,test_identity,on='TransactionID',how = 'left')
test_df
```

```python
test_df.isnull().sum()
```

```python
test_df['R_emaildomain'] = test_df['R_emaildomain'].fillna('Missing')
test_df['DeviceInfo'] = test_df['DeviceInfo'].fillna('Missing')
test_df['DeviceType'] = test_df['DeviceType'].fillna('Missing')
```

```python
empty_features_index = np.where(test_df.isnull().sum() > 0.5 * len(test_df))
empty_features_index 
```

```python
test_df.drop(columns = test_df.columns[empty_features_index],axis = 1 , inplace=True)
test_df
```

```python
null_value_cols = [cols for cols in test_df.columns if test_df[cols].isna().sum() > 0]
null_value_cols , len(null_value_cols)
```

```python
null_cat_cols = test_df[null_value_cols].select_dtypes(include='object').columns
null_float_cols = test_df[null_value_cols].select_dtypes(exclude='object').columns

null_cat_cols , null_float_cols
```

```python
for cols in null_cat_cols:
    print(f'{cols} = {test_df[cols].unique()}')
```

```python
for cols in null_cat_cols:
    test_df[cols] = test_df[cols].fillna(test_df[cols].mode()[0])
```

```python
for cols in null_cat_cols:
    print(f'{cols} = {test_df[cols].unique()}')
```

```python
for cols in null_float_cols:
    test_df[cols] = test_df[cols].fillna(test_df[cols].median())
    
test_df.isna().sum().sum()
```

```python
test_df.duplicated().sum()
```

```python
object_features = test_df.select_dtypes(include='object').columns
numerical_features = test_df.select_dtypes(exclude='object').columns

object_features , numerical_features
```

```python
for cols in object_features:
    print(f'{cols} = {test_df[cols].unique()}')
```

```python
for cols in numerical_features:
    print(f'{cols} = {test_df[cols].unique()}')
```

```python
test_df['Email_Match'] = (test_df['P_emaildomain'] == test_df['R_emaildomain']).astype(int)
test_df['P_email_Missing'] = (test_df['P_emaildomain']=='Missing').astype(int)
test_df['R_email_Missing'] = (test_df['R_emaildomain']=='Missing').astype(int)

test_df
```

```python
for cols in object_features:
    print(f'{cols} = {test_df[cols].unique()}')
```

```python
test_df['P_emaildomain'] = test_df['P_emaildomain'].map(test_df['P_emaildomain'].value_counts(normalize=True))
test_df['R_emaildomain'] = test_df['R_emaildomain'].map(test_df['R_emaildomain'].value_counts(normalize=True))

test_df['P_emaildomain']
```

```python
test_df['DeviceInfo'] = test_df['DeviceInfo'].map(test_df['DeviceInfo'].value_counts(normalize=True))
test_df['DeviceInfo']
```

```python
test_df.to_csv('naman_test.csv')
```
